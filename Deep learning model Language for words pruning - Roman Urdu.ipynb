```python
import pandas as pd
import string
```

```python
data = pd.read_csv("Roman Urdu DataSet.csv", header=None)
len(data[0])
```

```python
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)
roman_words_dataset = set()

for sent in data[0]:
    try:
        for word in sent.split():
            if hasNumbers(word.lower()):
                continue
            roman_words_dataset.add(word.translate(str.maketrans('', '', string.punctuation)).lower())
    except:
        continue
```

```python
len(roman_words_dataset)
```

```python
raw_text = ' '.join(list(roman_words_dataset))
raw_text
```

```python
# organize into sequences of characters
length = 8
sequences = list()
for i in range(length, len(raw_text)):
    # select sequence of tokens
    seq = raw_text[i-length:i+1]
    # store
    sequences.append(seq)
print('Total Sequences: %d' % len(sequences))
```

```python
sequences[:5]
```

```python
chars = sorted(list(set(raw_text)))
mapping = dict((c, i) for i, c in enumerate(chars))
```

```python
lines = sequences
sequences = list()

for line in lines:
	# integer encode line
	encoded_seq = [mapping[char] for char in line]
	# store
	sequences.append(encoded_seq)
```

```python
vocab_size = len(mapping)
print('Vocabulary Size: %d' % vocab_size)
```

```python
import numpy as np
sequences = np.array(sequences)
X, y = sequences[:,:-1], sequences[:,-1]
```

```python
from keras.utils import to_categorical
sequences = [to_categorical(x, num_classes=vocab_size) for x in X]
X = np.array(sequences)
y = to_categorical(y, num_classes=vocab_size)
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense, Activation
model = Sequential()
model.add(LSTM(75, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(vocab_size, activation='softmax'))
print(model.summary())
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit model
model.fit(X, y, epochs=100, verbose=2)
```
