```python
import os
from xml.dom import minidom
import string
from string import digits
import pandas as pd
import itertools
```

```python
dir_dataset = 'urdu dataset//COUNTER'
huroof_e_tahaji_urdu = set()
for file in os.listdir(dir_dataset):
    if file == '.DS_Store':
        continue
    mydoc = minidom.parse(dir_dataset+'//'+file)
    items = mydoc.getElementsByTagName('body')
    urdu_text = items[0].firstChild.nodeValue
    urdu_text = urdu_text.translate(str.maketrans('', '', string.punctuation))
    urdu_text = ''.join(list(filter(lambda x: x.isalpha(), urdu_text)))
    urdu_text = urdu_text.translate({ord(i): None for i in 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'})
    huroof_e_tahaji_urdu = huroof_e_tahaji_urdu.union(set(list(urdu_text)))
```

```python
dict_huroof_to_english_mapping = {}
for e, c in enumerate(huroof_e_tahaji_urdu):
    print('Current letter:', e, c, len(huroof_e_tahaji_urdu))
    dict_huroof_to_english_mapping[c] = list()
    n = input("Enter english letter or -1 to quit entering:")
    while n!='-1':
        dict_huroof_to_english_mapping[c].append(n)
        n = input("Enter roman letter(s) or -1 to quit entering:")
    print('-----------------------')
```

```python
with open('huroof mapping to english chars.txt', 'w') as f:
    for k in dict_huroof_to_english_mapping.keys():
        line = ''
        line+=k+','
        for c in dict_huroof_to_english_mapping[k]:
            line+=c+','
        line = line[:-1]
        print(k, line)
        f.write(line+'\n')

with open('huroof to decimal value mapping.txt','w') as f:
    for k in dict_huroof_to_english_mapping.keys():
        line = ''
        line+=k+','+k+'\n'
        f.write(line)
```

```python
dict_huroof_to_english_mapping = {}
with open('huroof mapping to english chars.txt') as f:
    for line in f:
        dict_huroof_to_english_mapping[line.split(',')[0].replace(u'\ufeff', '')] = [s.strip() for s in line.split(',')[1:]]
#dict_huroof_to_english_mapping
```

```python
def findOccurrences(s, ch):
    return [i for i, letter in enumerate(s) if letter == ch]
```

```python
def translate(word, dict_mapping):
    ambig_huroofs = []
    roman_word = ''
    for c, char in enumerate(word):
        #print(c, char)
        if char in dict_mapping.keys():
            if len(dict_mapping[char])>1:
                roman_word += '_'
                ambig_huroofs.append(dict_mapping[char])
            else:
                roman_word += dict_mapping[char][0]
                
    
    #print(roman_word)
    #print(ambig_huroofs)
    #print(list(itertools.product(*ambig_huroofs)))
    list_ambig_placements = findOccurrences(roman_word,'_')
    if roman_word.count('_')>1:
        for tuple_alteration in list(itertools.product(*ambig_huroofs)):
            roman_word_temp = list(roman_word)
            for k,t in enumerate(tuple_alteration):
                #print(k,t, list_ambig_placements[k])
                roman_word_temp[list_ambig_placements[k]] = t
            roman_word_temp = ''.join(roman_word_temp)
            ambig_huroofs.append(roman_word_temp)
        #print(ambig_huroofs)
    else:
        for tuple_alteration in list(itertools.product(*ambig_huroofs)):
            roman_word_temp = list(roman_word)
            for k, t in enumerate(tuple_alteration):
                roman_word_temp[list_ambig_placements[k]] = t
            roman_word_temp = ''.join(roman_word_temp)
            ambig_huroofs.append(roman_word_temp)
    #print(ambig_huroofs)
    ambig_words_only = [x for x in ambig_huroofs if type(x)!=list]
    print(ambig_words_only)
    return ambig_words_only
```
